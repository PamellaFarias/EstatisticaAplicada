(b) Crie uma função genérica para trocar duas colunas, sem codificar os nomes das colunas.

```python
 q8bdf = pd.DataFrame(np.arange(20).reshape(-1, 5), columns=list('abcde'))
```

```python
def trocar(q8bdf, col1=None, col2=None):
    fgenerica = q8adf.columns.tolist()
    b1, b2 = fgenerica.index(col1), fgenerica.index(col2)
    fgenerica[b2], fgenerica[b1] = fgenerica[b1], fgenerica[b2]
    return q8adf[fgenerica]
```

```python
q8bdf = trocar(q8bdf, 'a', 'c')
```
